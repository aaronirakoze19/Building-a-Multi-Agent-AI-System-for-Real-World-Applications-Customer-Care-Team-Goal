 Running the Multi-Agent Workflow

The cell above executes the supervisor, which coordinates the following agents:

- Greeter (Intent Classification)- This agent identifies the user’s intent by analyzing the input message and classifying it into categories in this scenarios  such as Refund/Billing.

- Researcher (Policy Retrieval) - this Retrieves relevant  information from the knowledge base (the customer care policy) and stores it in the state shared by other agents.

- Empath (Tone Adaptation) - This give the agent the ability to give a user friendly response with empathy to improve customer care .

- Resolver (Action Execution) - The Resolver decides what to actually do to fix the problem, this absically takes action to solve the issue.
